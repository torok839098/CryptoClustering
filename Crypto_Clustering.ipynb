```python
# Use the `StandardScaler()` module from scikit-learn to normalize the data from the CSV file

scaler = StandardScaler()

# Fit and transform the data
scaled_data = scaler.fit_transform(df_market_data)

# Print the mean and standard deviation before scaling
print("Before Scaling:")
print("Mean:", scaler.mean_)
print("Standard Deviation:", scaler.scale_)

# Print the scaled data
print("\nScaled Data:")
print(scaled_data)

```

---

```python
# Create a DataFrame with the scaled data


# Copy the crypto names from the original data

# Set the coinid column as index


# Display sample data

df_transformed_data=pd.DataFrame(scaled_data,columns=["price change percentage 24h","price change percentage 7d","price change percentage 14d","price change percentage 30d","price change percentage 60d","price change percentage 200d","price change percentage 1"])
df_transformed_data["coin_id"]=df_market_data.index
df_transformed_data = df_transformed_data.set_index("coin_id")
df_transformed_data.head(5)
```

```python
print(df_market_data.columns)
```

### Find the Best Value for k Using the Original Data.

```python
# Create a list with the number of k-values from 1 to 11
k = list(range(1,12))
k
```

```python
# Create an empty list to store the inertia values
inertia= []
```

```python
# Create a for loop to compute the inertia with each possible value of k
# Inside the loop:
# 1. Create a KMeans model using the loop counter for the n_clusters
# 2. Fit the model to the data using `df_market_data_scaled`
# 3. Append the model.inertia_ to the inertia list
inertia=[]
for i in k:
    k_model = KMeans(n_clusters=i, random_state=1)
    k_model.fit(df_transformed_data)
    inertia.append(k_model.inertia_)
```

```python
# Create a dictionary with the data to plot the Elbow curve
elbow_data= {"k": k, "inertia":inertia}
# Create a DataFrame with the data to plot the Elbow curve
df_elbow = pd.DataFrame(elbow_data)
df_elbow.head()
```

```python
# Plot a line chart with all the inertia values computed with 
# the different values of k to visually identify the optimal value for k.

k_original=df_elbow.hvplot.line(
    x="k", 
    y="inertia", 
    title="Elbow Curve", 
    xticks=k
)
k_original
```

#### Answer the following question: 

**Question:** What is the best value for `k`?

**Answer:** Based on the elbow method, the best value for 'k' is 4. Higher number of clusters makes the points closer to the centroid.

### Cluster Cryptocurrencies with K-means Using the Original Data

```python
# Initialize the K-Means model using the best value for k
model = KMeans(n_clusters=4, random_state=1)
model
```

```python
# Fit the K-Means model using the scaled data
```

```python
#model.fit(scaled_data) #Figuring out if it was scaled or transformed data 
```

```python
# Fit the K-Means model using the scaled data
model.fit(df_transformed_data)
```

```python
# Predict the clusters to group the cryptocurrencies using the scaled data


# Print the resulting array of cluster values.

crypto_clusters= model.predict(df_transformed_data)
print(crypto_clusters)
```

```python
# Create a copy of the DataFrame

cluster_predictions=df_transformed_data.copy()
```

```python
# Add a new column to the DataFrame with the predicted clusters
cluster_predictions['crypto_clusters']=crypto_clusters

# Display sample data
cluster_predictions.head()
```

```python
#verifying columns
cluster_predictions.columns
```

```python
# Create a scatter plot using hvPlot by setting 
# `x="price_change_percentage_24h"` and `y="price_change_percentage_7d"`. 
# Color the graph points with the labels found using K-Means and 
# add the crypto name in the `hover_cols` parameter to identify 
# the cryptocurrency represented by each data point.
cluster_predictions_plot=cluster_predictions.hvplot.scatter(
    x='price change percentage 24h',
    y='price change percentage 7d',
    by='crypto_clusters',
    hover_cols= 'coin_id' 
)
cluster_predictions_plot
```

### Optimize Clusters with Principal Component Analysis.

```python
# Create a PCA model instance and set `n_components=3`.
pca = PCA(n_components=3)
```

```python
# Use the PCA model with `fit_transform` to reduce to 
# three principal components.

# View the first five rows of the DataFrame. 
cluster_predictions_pca = pca.fit_transform(cluster_predictions)
cluster_predictions_pca[0:5]
```

```python
# Retrieve the explained variance to determine how much information 
# can be attributed to each principal component.
pca.explained_variance_ratio_
```

```python
# Total explained variance 
sum(pca.explained_variance_ratio_)
```

#### Answer the following question: 

**Question:** What is the total explained variance of the three principal components?

**Answer:** The total explained variance of the three principal components is 0.8886218549859446 or about 89%. 

```python
# Create a new DataFrame with the PCA data.
# Creating a DataFrame with the PCA data
cluster_predictions_pca_df = pd.DataFrame(cluster_predictions_pca,
                                          columns= ["PCA1", "PCA2","PCA3"]
                                          
)
cluster_predictions_pca_df.head()
```

```python
# Copy the crypto names from the original data
# Set the coinid column as index
# Display sample data
#cluster_predictions_pca_df=pd.DataFrame(scaled_data,columns=["price change percentage 24h","price change percentage 7d","price change percentage 14d","price change percentage 30d","price change percentage 60d","price change percentage 200d","price change percentage 1"])
cluster_predictions_pca_df["coin_id"]=cluster_predictions.index
cluster_predictions_pca_df = cluster_predictions_pca_df.set_index("coin_id")
cluster_predictions_pca_df.head()
```

---

### Find the Best Value for k Using the PCA Data

```python
# Create a list with the number of k-values from 1 to 11
k = list(range(1,12))
k
```

```python
# Create an empty list to store the inertia values


# Create a for loop to compute the inertia with each possible value of k
# Inside the loop:
# 1. Create a KMeans model using the loop counter for the n_clusters
# 2. Fit the model to the data using `df_market_data_pca`
# 3. Append the model.inertia_ to the inertia list

inertia=[]
for i in k:
    k_model = KMeans(n_clusters=i, random_state=1)
    k_model.fit(cluster_predictions_pca_df)
    inertia.append(k_model.inertia_)
```

```python
# Create a dictionary with the data to plot the Elbow curve

# Create a DataFrame with the data to plot the Elbow curve
elbow_data= {"k": k, "inertia":inertia}
df_elbow = pd.DataFrame(elbow_data)
df_elbow.head()
```

```python
# Plot a line chart with all the inertia values computed with 
# the different values of k to visually identify the optimal value for k.



pca_data_plot=df_elbow.hvplot.line(
    x="k", 
    y="inertia", 
    title="Elbow Curve", 
    xticks=k
)
pca_data_plot
```

#### Answer the following questions: 

* **Question:** What is the best value for `k` when using the PCA data?

  * **Answer:** The best value for k by using PCA is 4.


* **Question:** Does it differ from the best k value found using the original data?

  * **Answer:** Both are ideal by using 4 clusters, however the inertia values of each are different. 

### Cluster Cryptocurrencies with K-means Using the PCA Data

```python
# Initialize the K-Means model using the best value for k
model = KMeans(n_clusters=4, random_state=0)
```

```python
# Fit the K-Means model using the PCA data
model.fit(cluster_predictions_pca_df)
```

```python
# Predict the clusters to group the cryptocurrencies using the PCA data

# Print the resulting array of cluster values.

k_4 = model.predict(cluster_predictions_pca_df)
print(k_4)
```

```python
# Create a copy of the DataFrame with the PCA data
cluster_pca_predictions_df = cluster_predictions_pca_df.copy()
```

```python
# Add a new column to the DataFrame with the predicted clusters
cluster_pca_predictions_df["cluster_segments"] = k_4
```

```python
# Display sample data
cluster_pca_predictions_df.head()
```

```python
# Create a scatter plot using hvPlot by setting 
# `x="PC1"` and `y="PC2"`. 
# Color the graph points with the labels found using K-Means and 
# add the crypto name in the `hover_cols` parameter to identify 
# the cryptocurrency represented by each data point.



cluster_pca_predictions_plot=cluster_pca_predictions_df.hvplot.scatter(
    x="PCA1",
    y="PCA2",
    by="cluster_segments",
    hover_cols="coin_id"
)
cluster_pca_predictions_plot
```
